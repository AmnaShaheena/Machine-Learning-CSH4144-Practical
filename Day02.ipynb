```python
m=np.array([[1,5,2],
          [3,7,4],
          [2,0,9]])
```

```python
print(m)
```

```python
print("Matrix Transpose",m.transpose(),'\n')
```

```python
print('Matrix Determinant',np.linalg.det(m),'\n')
```

```python
m_inv=np.linalg.inv(m)
print('Matrix inverse',m_inv,'\n')
```

```python
iden_m=np.dot(m,m_inv)
iden_m=np.round(np.abs(iden_m),0)
print('Product of matrix and its inverse:\n',iden_m)
```

```python
eigen_vals,eigen_vecs=np.linalg.eig(m)
print('Eigen Values',eigen_vals,'\n')
print('Eigen Vectors',eigen_vecs,'\n' )
```

```python
U,S,VT=np.linalg.svd(m)
print('getting SVD outputs:-\n')
print('U:\n',U,'\n')
print('S:\n',S,'\n')
print('VT:\n',VT,'\n')
```

# Random Variables

```python
#descriptive statistics
import scipy as sp
import numpy as np
```

```python
#get data
nums=np.random.randint(1,20,size=(1,15))[0]
print('Data: ',nums)
```

```python
#get descriptive stats
print('Mean:',np.mean(nums))
print('Median:',np.median(nums))
print('Standard Deviation:',np.std(nums))
print('Variance:',np.var(nums))

print('Mode:',sp.stats.mode(nums))
print('Skewness:',sp.stats.skew(nums))
print('Kurtosis:',sp.stats.kurtosis(nums))
```

Q2)Write a code to generate 100 lottery ticket and pick two lucky tickets from winner
  The lottery number must be 10 digits long
  All 100 ticket number must be unique
Generate a random list of 100 numbers using randrange() and then use the sample() method to pick lucky tickets

```python
import random

# Generate 100 unique 10-digit lottery numbers
lottery_tickets = random.sample(range(10**9, 10**10), 100)

# Pick two lucky tickets
lucky_tickets = random.sample(lottery_tickets, 2)

print("Generated lottery tickets:", lottery_tickets)
print("Lucky tickets:", lucky_tickets)
```

Q3)Write a code to generate a random password which meets the following conditions
    password length must be 10 characters long
    it must contain at least upper case Letters, 1 digit and 1 special symbol

```python
import random
import string

def generate_password(length=10):
    """Generates a random password with at least one uppercase letter, one digit, and one special character."""
    if length < 3:
        return "Password length must be at least 3 characters."

    # Define character sets
    uppercase_letters = string.ascii_uppercase
    digits = string.digits
    special_characters = string.punctuation
    all_characters = uppercase_letters + digits + special_characters + string.ascii_lowercase

    # Ensure at least one of each required character type
    password = [
        random.choice(uppercase_letters),
        random.choice(digits),
        random.choice(special_characters)
    ]

    # Fill the rest of the password length
    password += random.choices(all_characters, k=length - 3)

    # Shuffle the password list to make it random
    random.shuffle(password)

    # Join the list into a string
    return "".join(password)

# Generate and print a password
password = generate_password()
print("Generated password:", password)
```
